```python
#creating the Hangman game 
import random as rd
from hangman_art import stages, logo, word_list

choice_list = rd.choice(word_list)
length = len(choice_list)
end_game = False
lives = 6

print(logo)

# Creating blanks
display = ["_" for _ in range(length)]

while not end_game:
    guess = input("Guess a letter: ").lower()

    # Check if the guessed letter is in the word
    if guess in display:
        print(f"You've already guessed the letter {guess}.")

    # Check guessed letter
    for position in range(length):
        letter = choice_list[position]
        if letter == guess:
            display[position] = letter
            print(display)

    # Check if the guessed letter is not in the word
    if guess not in choice_list:
        lives -= 1
        print(stages[lives]) 
        if lives == 0:
            end_game = True
            print("You lose. The correct word was:", choice_list)
            print(stages[lives]) 

    # Display the current state of the word
    print(" ".join(display))

    # Check if the player guessed all letters
    if "_" not in display:
        end_game = True
        print("Congratulations! You win.")
```
